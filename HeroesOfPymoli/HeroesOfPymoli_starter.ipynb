```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

```python
totalplayers_df=len(purchase_data["SN"].unique())
total_players=pd.DataFrame({"Total Players":[totalplayers_df]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
NumUniqueItems=len(purchase_data["Item Name"].unique())
AvgPrice=purchase_data["Price"].mean()
NumofPurch=len(purchase_data["Item Name"])
TotalRevenue=purchase_data["Price"].sum()

Purchasing_Analysis=pd.DataFrame({"Number of Unique Item":[NumUniqueItems],"Average Price":[AvgPrice],"Number of Purchases":[NumofPurch],"Total Revenue":[TotalRevenue]})
Purchasing_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
trueSN=purchase_data.drop_duplicates(subset='SN',keep="first")
Total=trueSN["Gender"].count()
TotalMale=trueSN["Gender"].value_counts()['Male']
TotalFemale=trueSN["Gender"].value_counts()['Female']
TotalOther=trueSN["Gender"].value_counts()['Other / Non-Disclosed']

MalePercentage=(TotalMale/Total)*100
FemalePercentage=(TotalFemale/Total)*100
OtherPercentage=(TotalOther/Total)*100

GenderDemographics=pd.DataFrame({"":["Male","Female","Other/Non-Disclosed"],"Total Count":[TotalMale,TotalFemale, TotalOther],"Percentage of Players":[MalePercentage, FemalePercentage, OtherPercentage]})
GenderDemographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
genderdata=purchase_data.groupby(["Gender"])
PurchCount=genderdata["SN"].count()
AvgPrice=genderdata["Price"].mean()
TotalValue=genderdata["Price"].sum()
AvgTotal=(genderdata["Price"].sum()/genderdata["SN"].count())

PurchasingAnalysis=pd.DataFrame({"Purchase Count":PurchCount,"Average Purchase Price":AvgPrice,"Total Purchase Value":TotalValue, "Avg Total Purchase per Person":AvgTotal})
PurchasingAnalysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,10,15,20,25,30,35,40,200]
binLab = ['<10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', '40+']
demographics=purchase_data.copy()
pd.cut(demographics["Age"], bins, labels=binLab)
demographics["Age Group"]=pd.cut(demographics["Age"], bins, labels=binLab)
AgeDemographics = demographics.groupby(["SN","Age Group"])
players = pd.DataFrame(AgeDemographics.size())
AgeDemographics = pd.DataFrame(players.groupby(["Age Group"]).count())
AgeDemographics.columns=["Total Count"]
AgeDemographics["Percentage of Payers"]=round(100*AgeDemographics["Total Count"]/AgeDemographics["Total Count"].sum(),2)
AgeDemographics


```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SNList=purchase_data.groupby(["SN"])
PurchCount=SNList["Item ID"].count()
TotalPrice=SNList["Price"].sum()
AvgPurch=(TotalPrice/PurchCount)

TopSpenders=pd.DataFrame({"Purchase Count": PurchCount,"Average Purchase Price":AvgPurch,"Total Purchase Value":TotalPrice})
TopSpenders=TopSpenders.sort_values("Total Purchase Value",ascending=False)
TopSpenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
PopularItems=purchase_data.groupby(["Item ID","Item Name"])
PurchCount=PopularItems["SN"].count()
IPrice=PopularItems["Price"].sum()
AvgPrice=(IPrice/PurchCount)
Value=(AvgPrice*PurchCount)

MostPopularItems=pd.DataFrame({"Purchase Count":PurchCount, "Item Price":AvgPrice, "Total Purchase Value":Value})
MostPopularItems=MostPopularItems.sort_values("Purchase Count", ascending=False)
MostPopularItems.head()
```

## Most Profitable Items
